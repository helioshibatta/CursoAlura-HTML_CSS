```python
import pandas as pd
```

```python
candidatos2022 = pd.read_csv('/content/Lista-Candidatos-Deputado Distrital-DISTRITO FEDERAL-2022.csv', delimiter=";")
```

```python
candidatos2022['Nome Urna'] = candidatos2022['Nome Urna'].str.strip()
candidatos2022['Nome Completo'] = candidatos2022['Nome Completo'].str.strip()
candidatos2022.head()
```

```python
candidatos2018_2 = pd.read_csv('/content/candidatos2018.csv', delimiter=";")
candidatos2018_2['Nome Urna'] = candidatos2018_2['Nome Urna'].str.strip()
candidatos2018_2['Nome Urna'] = candidatos2018_2['Nome Urna'].str.upper()
candidatos2018_2.head()
```

```python
candidatos2018_1 = pd.read_csv('/content/Lista-Candidatos-Deputado Distrital-DISTRITO FEDERAL-2018.csv', delimiter=";")
candidatos2018_1['Nome Urna'] = candidatos2018_1['Nome Urna'].str.upper()
candidatos2018_1['Nome Urna'] = candidatos2018_1['Nome Urna'].str.strip()
candidatos2018_1['Nome Completo'] = candidatos2018_1['Nome Completo'].str.strip()
candidatos2018_1.head()
```

```python
candidatos2018 = pd.merge(candidatos2018_1, candidatos2018_2,on="Nome Urna")
candidatos2018.head()
```

```python
candidatos2018['Nome Urna'] = candidatos2018['Nome Urna'].str.upper()
candidatos2018['Nome Urna'] = candidatos2018['Nome Urna'].str.strip()
candidatos2018['Nome Completo'] = candidatos2018['Nome Completo'].str.strip()
#parâmetro inner
tabela_votos = pd.merge(candidatos2022, candidatos2018, on='Nome Completo', how='left')

#tabela_votos = pd.merge(candidatos2022, candidato2018, left_on='Nome Urna', right_on='Nome Urna')  caso os campos sejam diferentes
```

```python
candidatos2014 = pd.read_csv('/content/candidatos2014.csv', delimiter=";")

candidatos2014['Nome Urna'] = candidatos2014['Nome Urna'].str.strip()
candidatos2014.head()
```

```python
candidatos2014_2 = pd.read_csv('/content/merge20202018.csv', delimiter=",")
candidatos2014_2.head()
```

```python
candidatos2014_2['Nome Urna_x'] = candidatos2014_2['Nome Urna_x'].str.strip()
candidatos2014['Nome Urna'] = candidatos2014['Nome Urna'].str.strip()

tabela_votos = pd.merge(candidatos2014_2, candidatos2014, left_on='Nome Urna_x', right_on='Nome Urna', how='left')
tabela_votos.head()
```
